```python
import sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import precision_recall_curve, f1_score

data = test

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.drop("Label", axis=1), data["Label"], test_size=0.3, random_state=123)

# Train and evaluate logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
probs_logreg = logreg.predict_proba(X_test)[:, 1]
precision_logreg, recall_logreg, thresholds_logreg = precision_recall_curve(y_test, probs_logreg)
f1_scores_logreg = 2 * (precision_logreg * recall_logreg) / (precision_logreg + recall_logreg)

# Train and evaluate random forest model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
probs_rf = rf.predict_proba(X_test)[:, 1]
precision_rf, recall_rf, thresholds_rf = precision_recall_curve(y_test, probs_rf)
f1_scores_rf = 2 * (precision_rf * recall_rf) / (precision_rf + recall_rf)

# Train and evaluate Naive Bayes model
nb = MultinomialNB()
nb.fit(X_train, y_train)
probs_nb = nb.predict_proba(X_test)[:, 1]
precision_nb, recall_nb, thresholds_nb = precision_recall_curve(y_test, probs_nb)
f1_scores_nb = 2 * (precision_nb * recall_nb) / (precision_nb + recall_nb)

# Find the optimal threshold value for each model
optimal_idx_logreg = np.argmax(f1_scores_logreg)
optimal_threshold_logreg = thresholds_logreg[optimal_idx_logreg]

optimal_idx_rf = np.argmax(f1_scores_rf)
optimal_threshold_rf = thresholds_rf[optimal_idx_rf]

optimal_idx_nb = np.argmax(f1_scores_nb)
optimal_threshold_nb = thresholds_nb[optimal_idx_nb]

# Determine the most optimal threshold value
optimal_threshold = (optimal_threshold_logreg + optimal_threshold_rf + optimal_threshold_nb) / 3

# Set the threshold value for each model
logreg_threshold = optimal_threshold / 2
rf_threshold = optimal_threshold / 2
nb_threshold = optimal_threshold / 2

print(logreg_threshold, rf_threshold, nb_threshold )
```
